```python
import pandas as pd
import numpy as np
import re
from scipy import stats
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.feature_extraction import DictVectorizer
```

```python
pd.options.display.max_seq_items = 2000
```

```python
pd.options.display.max_rows = 2000
pd.options.display.max_columns = 2000
```

```python
airbnb_df = pd.read_csv('merged_dataset.csv')
```

##### Add Zipcode Latitude and Longitude
***

```python
zipcode = pd.read_csv('us-zip-code-latitude-and-longitude.csv', sep=';')
```

```python
zc_df = zipcode[zipcode.Zip.isin(airbnb_df.zipcode.unique())]
```

```python
zc = ['Zip', 'Latitude', 'Longitude']
zc_df = zc_df[zc]
```

```python
airbnb_df = airbnb_df.merge(zc_df, 
                            left_on = 'zipcode', 
                            right_on = 'Zip')
airbnb_df.drop(columns = ['Zip'], inplace = True)
```

```python
airbnb_df.head()
```

##### Selecting some features that seem important
***

```python
airbnb_col = ['zipcode', 'property_type', 
              'room_type', 'accommodates', 'bathrooms', 
              'bedrooms', 'beds', 'bed_type', 'neighbourhood_cleansed',
              'amenities', 'price', 'cleaning_fee', 'guests_included', 'extra_people', 
              'number_of_reviews', 'review_scores_rating', 
              'Latitude', 'Longitude']
```

```python
airbnb_df = airbnb_df[airbnb_col]
```

```python
airbnb_df.rename(columns={'Latitude': 'zc_lat', 'Longitude': 'zc_lon'}, 
                 inplace = True)
```

#### Dropping rows with NA values
***

```python
airbnb_df = airbnb_df.dropna().reset_index(drop=True)
airbnb_df.shape
```

##### Monetary Value and Floats to Int
***

```python
money = ['price', 'cleaning_fee', 'extra_people']
```

```python
for m in money:
    airbnb_df[m].fillna('$0.00', inplace = True)
```

```python
airbnb_df['price'] = airbnb_df.price.str[1:-3]
airbnb_df['price'] = airbnb_df.price.str.replace(r'[^0-9]+', '')
airbnb_df['price'] = airbnb_df.price.astype(int)
```

```python
airbnb_df['cleaning_fee'] = airbnb_df.cleaning_fee.str[1:-3]
airbnb_df['cleaning_fee'] = airbnb_df.cleaning_fee.str.replace(r'[^0-9]+', '')
airbnb_df['cleaning_fee'] = airbnb_df.cleaning_fee.astype(int)
```

```python
airbnb_df['extra_people'] = airbnb_df.extra_people.str[1:-3]
airbnb_df['extra_people'] = airbnb_df.extra_people.str.replace(r'[^0-9]+', '')
airbnb_df['extra_people'] = airbnb_df.extra_people.astype(int)
```

```python
airbnb_df['bathrooms'] = airbnb_df['bathrooms'].astype(int)
airbnb_df['beds'] = airbnb_df['beds'].astype(int)
airbnb_df['bedrooms'] = airbnb_df['bedrooms'].astype(int)
```

##### Label Encoder
***

```python
le = LabelEncoder()
```

```python
p_type = airbnb_df['property_type'].unique()
le.fit(p_type)
airbnb_df['property_type'] = le.transform(airbnb_df.property_type)
```

```python
r_type = airbnb_df['room_type'].unique()
le.fit(r_type)
airbnb_df['room_type'] = le.transform(airbnb_df.room_type)
```

```python
b_type = airbnb_df['bed_type'].unique()
le.fit(b_type)
airbnb_df['bed_type'] = le.transform(airbnb_df.bed_type)
```

```python
n_cleansed = airbnb_df['neighbourhood_cleansed'].unique()
le.fit(n_cleansed)
airbnb_df['neighbourhood_cleansed'] = le.transform(airbnb_df.neighbourhood_cleansed)
```

##### Amenities One-Hot Encoder
***

```python
a_cleansed = []
for a in airbnb_df.amenities:
    a_cleansed.append(re.sub(r'[\"\{\}]+', '', a.lower()).split(','))

airbnb_df['amenities'] = a_cleansed
```

```python
airbnb_df['num_amenities'] = airbnb_df.amenities.apply(lambda x: len(x))
```

```python
list_amenities = []

for a in airbnb_df.amenities:
    for am in a:
        list_amenities.append(am)

list_amenities = np.unique(list_amenities)[1:]
```

```python
# list_amenities
```

```python
necessary_amenities = ['air conditioning', 'cat(s)', 'dog(s)',
                       'dryer', 'heating', 'internet',
                       'kitchen', 'microwave',
                       'outdoor parking', 'private bathroom',
                       'self check-in', 'tv', 'washer', 'wifi']
```

```python
for a in necessary_amenities:
    airbnb_df[a] = airbnb_df.amenities.apply(lambda x: a in x).astype(int)
```

```python
# airbnb_df.head()
```

##### Dropping outliers based
***

```python
z = np.abs(stats.zscore(airbnb_df['price']))
threshold = 3
drop_list = np.where(z > 3)
drop_list = drop_list[0]
```

```python
airbnb_df = airbnb_df.drop(drop_list, axis=0)
```

```python
x = airbnb_df.sort_values(by=['price'])
x = x.reset_index(drop=True)
x['index'] = x.index
```

```python
x.plot.scatter(x='index', y='price')
```

#### Dropping amenities and zipcode since we performed feature engineering on them to produce new features
***

```python
airbnb_df = airbnb_df.drop(['zipcode' , 'amenities'], axis=1)
```

***
